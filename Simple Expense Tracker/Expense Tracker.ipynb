```python

import time 
import openpyxl
import numpy as np
from openpyxl.utils import get_column_letter
from openpyxl.styles import Font
from openpyxl.styles import Alignment
from tkinter import *
from tkinter import messagebox
start_time = time.time()
def get_numeric_value(entry_widget):
    try:
        value = float(entry_widget.get())
        return value
    except ValueError:
        return None
def add_expense(date,food,fa,grocery,ga,shopping,sa,entertainment,ea,others,oa):
    wb = openpyxl.load_workbook(excel_file)
    sheet = wb.active
    row = (date,food,fa,grocery,ga,shopping,sa,entertainment,ea,others,oa)
    sheet.append(row)
    wb.save(excel_file)
    wb.close()
    messagebox.showinfo("Expense Added", "Expense added successfully!")
    


def myClick1():
    response1, response2 = food_exp()
    new_window = Toplevel(root)
    new_window.title("Food Expense Details")
    # Display the information in the new window
    mylabel1 = Label(new_window, text="Record of Food ")
    mylabel1.pack()
    mylabel2 = Label(new_window, text=f"Expense on each Item Rs.\n{response1}", font=20,fg="red",padx=100,pady=100)
    mylabel2.pack()
    mylabel3 = Label(new_window, text=f"Total Expense Rs. : {response2}", font=20,fg="red",padx=100,pady=100)
    mylabel3.pack()

def myClick2():
    response1, response2 = groc_exp()
    new_window = Toplevel(root)
    new_window.title("Grocery Expense Details")
    # Display the information in the new window
    mylabel1 = Label(new_window, text="Record of Grocery ")
    mylabel1.pack()
    mylabel2 = Label(new_window, text=f"Expense on each Item Rs.\n{response1}", font=10,fg="red",padx=80,pady=80)
    mylabel2.pack()
    mylabel3 = Label(new_window, text=f"Total Expense Rs. : {response2}", font=10,fg="red",padx=80,pady=80)
    mylabel3.pack()

def myClick3():
    response1, response2 = shop_exp()
    new_window = Toplevel(root)
    new_window.title("Shopping Expense Details")
    # Display the information in the new window
    mylabel1 = Label(new_window, text="Detail of Grocery ")
    mylabel1.pack()
    mylabel2 = Label(new_window, text=f"Expense on each Item Rs.\n{response1}", font=10,fg="red",padx=80,pady=80)
    mylabel2.pack()
    mylabel3 = Label(new_window, text=f"Total Expense Rs. : {response2}", font=10,fg="red",padx=80,pady=80)
    mylabel3.pack()

def myClick4():
    response1, response2 = ent_exp()
    new_window = Toplevel(root)
    new_window.title("Entertainment Expense Details")
    # Display the information in the new window
    mylabel1 = Label(new_window, text="Detail of Entertainment ")
    mylabel1.pack()
    mylabel2 = Label(new_window, text=f"Expense on each Item Rs.\n{response1}", font=10,fg="red",padx=80,pady=80)
    mylabel2.pack()
    mylabel3 = Label(new_window, text=f"Total Expense Rs. : {response2}", font=10,fg="red",padx=80,pady=80)
    mylabel3.pack()

def myClick5():
    response1, response2 = oth_exp()
    new_window = Toplevel(root)
    new_window.title("Other Expenses Details")
    # Display the information in the new window
    mylabel1 = Label(new_window, text="Detail of Other Items ")
    mylabel1.pack()
    mylabel2 = Label(new_window, text=f"Expense on each Item Rs.\n{response1}", font=10,fg="red",padx=80,pady=80)
    mylabel2.pack()
    mylabel3 = Label(new_window, text=f"Total Expense Rs. : {response2}", font=10,fg="red",padx=80,pady=80)
    mylabel3.pack()

def myClick6():
    response1 = total_expense()
    new_window = Toplevel(root)
    new_window.title("Overall Expenses Amount")
    mylabel1 = Label(new_window, text=f"Overall Expense Rs.\n{response1}", font=20,fg="red",padx=100,pady=100)
    mylabel1.pack()

root = Tk()
root.title("Expense Tracker")
heading=Label(text="Simple Expense Tracker",pady=40,font=("Helvetica", 20), fg="blue")
heading.pack()
# Create a frame for buttons
frame_buttons = LabelFrame(root, text="Expense Categories", padx=30, pady=30)
frame_buttons.pack(pady=10)

b1 = Button(frame_buttons, text='Food', padx=20, pady=10, command=myClick1)
b2 = Button(frame_buttons, text='Grocery', padx=20, pady=10, command=myClick2)
b3 = Button(frame_buttons, text='Shopping', padx=20, pady=10, command=myClick3)
b4 = Button(frame_buttons, text='Entertainment', padx=20, pady=10, command=myClick4)
b5 = Button(frame_buttons, text='Other Items', padx=20, pady=10, command=myClick5)
b6 = Button(frame_buttons, text='Overall Expense', padx=20, pady=10, command=myClick6)

b1.grid(row=0, column=0, padx=10, pady=10)
b2.grid(row=0, column=1, padx=10, pady=10)
b3.grid(row=0, column=2, padx=10, pady=10)
b4.grid(row=1, column=0, padx=10, pady=10)
b5.grid(row=1, column=1, padx=10, pady=10)
b6.grid(row=1, column=2, padx=10, pady=10)

# Create a frame for adding expenses
frame_add_expense = LabelFrame(root, text="Add Expense", padx=50)
frame_add_expense.pack(padx=50, pady=10)

# Labels and entry widgets
Label(frame_add_expense, text="Date:").grid(row=0, column=0, padx=5, pady=5)
date_entry = Entry(frame_add_expense)
date_entry.grid(row=0, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Food Detail:").grid(row=1, column=0, padx=5, pady=5)
food_entry = Entry(frame_add_expense)
food_entry.grid(row=1, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Food Amount:").grid(row=2, column=0, padx=5, pady=5)
fae = Entry(frame_add_expense)
fae.grid(row=2, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Grocery Detail:").grid(row=3, column=0, padx=5, pady=5)
groc_entry = Entry(frame_add_expense)
groc_entry.grid(row=3, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Grocery Amount:").grid(row=4, column=0, padx=5, pady=5)
gae = Entry(frame_add_expense)
gae.grid(row=4, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Shopping Detail:").grid(row=5, column=0, padx=5, pady=5)
shop_entry = Entry(frame_add_expense)
shop_entry.grid(row=5, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Shopping Amount:").grid(row=6, column=0, padx=5, pady=5)
sae = Entry(frame_add_expense)
sae.grid(row=6, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Entertainment Detail:").grid(row=7, column=0, padx=5, pady=5)
ent_entry = Entry(frame_add_expense)
ent_entry.grid(row=7, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Entertainment Amount:").grid(row=8, column=0, padx=5, pady=5)
eae = Entry(frame_add_expense)
eae.grid(row=8, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Other items Detail:").grid(row=9, column=0, padx=5, pady=5)
oth_entry = Entry(frame_add_expense)
oth_entry.grid(row=9, column=1, padx=5, pady=5)

Label(frame_add_expense, text="Other items Amount:").grid(row=10, column=0, padx=5, pady=5)
oae = Entry(frame_add_expense)
oae.grid(row=10, column=1, padx=5, pady=5)

# Button to add expenses
add_button = Button(frame_add_expense, text="Add Expense", command=lambda: add_expense(
    date_entry.get(), food_entry.get(), get_numeric_value(fae),
    groc_entry.get(), get_numeric_value(gae),
    shop_entry.get(), get_numeric_value(sae),
    ent_entry.get(), get_numeric_value(eae),
    oth_entry.get(), get_numeric_value(oae)
))
add_button.grid(row=11, column=0, columnspan=2, pady=10)

root.mainloop()





# Define the Excel file path
excel_file = 'expense_tracker.xlsx'


# Function to add an expense
# def add_expense(date,food,fa,grocery,ga,shopping,sa,entertainment,ea,others,oa):
#     wb = openpyxl.load_workbook(excel_file)
#     sheet = wb.active
#     row = (date,food,fa,grocery,ga,shopping,sa,entertainment,ea,others,oa)
#     sheet.append(row)
#     wb.save(excel_file)
#     wb.close()

# Function to see amount on food
def food_exp():
    wb2=openpyxl.load_workbook(excel_file)
    sheet2=wb2['Expenses']
    col1=1
    col2=2
    data=[]
    data2=[]
    for row in sheet2.iter_rows(min_col=col2 + 1, max_col=col2 + 1):
        for cell in row:
            values=cell.value
            data.append(values)
    del data[0]
    arr1=np.array([data])
    sum_of_val1=np.sum(arr1)
    for row2 in sheet2.iter_rows(min_col=col1 + 1, max_col=col1 + 1):
        for cell2 in row2:
            values2=cell2.value
            data2.append(values2)
    del data2[0]
    arr1=np.array([[data2],[data]])
    trans=np.transpose(arr1)
    return trans,sum_of_val1



def groc_exp():
    wb2=openpyxl.load_workbook(excel_file)
    sheet2=wb2['Expenses']
    col1=3
    col2=4
    data=[]
    data2=[]
    for row in sheet2.iter_rows(min_col=col2 + 1, max_col=col2 + 1):
        for cell in row:
            values=cell.value
            data.append(values)
    del data[0]
    arr1=np.array([data])
    sum_of_val1=np.sum(arr1)
    for row2 in sheet2.iter_rows(min_col=col1 + 1, max_col=col1 + 1):
        for cell2 in row2:
            values2=cell2.value
            data2.append(values2)
    del data2[0]
    arr1=np.array([[data2],[data]])
    trans=np.transpose(arr1)
    return trans,sum_of_val1



    
def shop_exp():
    wb2=openpyxl.load_workbook(excel_file)
    sheet2=wb2['Expenses']
    col1=5
    col2=6
    data=[]
    data2=[]
    for row in sheet2.iter_rows(min_col=col2 + 1, max_col=col2 + 1):
        for cell in row:
            values=cell.value
            data.append(values)
    del data[0]
    arr1=np.array([data])
    sum_of_val1=np.sum(arr1)
    for row2 in sheet2.iter_rows(min_col=col1 + 1, max_col=col1 + 1):
        for cell2 in row2:
            values2=cell2.value
            data2.append(values2)
    del data2[0]
    arr1=np.array([[data2],[data]])
    trans=np.transpose(arr1)
    #print(trans)
    #print("Total Expenditure on Shopping: ",sum_of_val1)
    return trans,sum_of_val1
    
    
def ent_exp():
    wb2=openpyxl.load_workbook(excel_file)
    sheet2=wb2['Expenses']
    col1=7
    col2=8
    data=[]
    data2=[]
    for row in sheet2.iter_rows(min_col=col2 + 1, max_col=col2 + 1):
        for cell in row:
            values=cell.value
            data.append(values)
    del data[0]
    arr1=np.array([data])
    sum_of_val1=np.sum(arr1)
    for row2 in sheet2.iter_rows(min_col=col1 + 1, max_col=col1 + 1):
        for cell2 in row2:
            values2=cell2.value
            data2.append(values2)
    del data2[0]
    arr1=np.array([[data2],[data]])
    trans=np.transpose(arr1)
    #print(trans)
    #print("Total Expenditure on Entertainment: ",sum_of_val1)
    return trans,sum_of_val1
    
    
def oth_exp():
    wb2=openpyxl.load_workbook(excel_file)
    sheet2=wb2['Expenses']
    col1=9
    col2=10
    data=[]
    data2=[]
    for row in sheet2.iter_rows(min_col=col2 + 1, max_col=col2 + 1):
        for cell in row:
            values=cell.value
            data.append(values)
    del data[0]
    arr1=np.array([data])
    sum_of_val1=np.sum(arr1)
    for row2 in sheet2.iter_rows(min_col=col1 + 1, max_col=col1 + 1):
        for cell2 in row2:
            values2=cell2.value
            data2.append(values2)
    del data2[0]
    arr1=np.array([[data2],[data]])
    trans=np.transpose(arr1)
    #print(trans)
    #print("Total Expenditure on other items: ",sum_of_val1)
    return trans,sum_of_val1
    
    
def total_expense():
    
    te2_sum = 0

    _, te2 = food_exp()
    te2_sum += te2

    _, te2 = groc_exp()
    te2_sum += te2

    _, te2 = shop_exp()
    te2_sum += te2

    _, te2 = ent_exp()
    te2_sum += te2

    _, te2 = oth_exp()
    te2_sum += te2

    return te2_sum
# Function to create the expense tracker if it doesn't exist

def create_expense_tracker():
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet.title = "Expenses"

    # Set column headers
    headers = ['Date','Food','Amount Spent on Food','Grocery','Amount Spent on Grocery','Shopping','Amount Spent on Shopping','Entertainment','Amount Spent on Entertainment','Others','Amount Spent on Others']
    for col_num, header in enumerate(headers, 1):
        cell = sheet.cell(row=1, column=col_num)
        cell.value = header
        cell.font = Font(bold=True)
        cell.alignment = Alignment(horizontal='center')

    wb.save(excel_file)
    wb.close()

# Check if the expense tracker exists; if not, create it
try:
    wb = openpyxl.load_workbook(excel_file)
except FileNotFoundError:
    create_expense_tracker()
    
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Total computation time: {elapsed_time} seconds")
    
    
    
#Total computation time: 2.019258737564087 seconds (before performaing any operations)    
#Total computation time: 93.24021983146667 seconds (After performing operations)
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    


```
